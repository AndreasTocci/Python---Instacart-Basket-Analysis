# Question 4

Follow the instructions in the Exercise for creating a loyalty flag for existing customers using the transform() and loc() functions.

## Transform

```python
#1.Split the data into groups based on the “user_id” column.

#2.Apply the transform() function on the “order_number” column to generate the maximum orders for each user.

#3. Create a new column, “max_order,” into which it will place the results of your aggregation.

ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
# check output
ords_prods_merge.head()
```

## Creating Loyalty Flag for Transformed data

```python
# locate max order in ords_prods_merge if its greater than 40 in column loyalty_flag return loyal customer
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal Customer'

# locate max order in ords_prods_merge if its <= 40  and > 10 in column loyalty_flag return regular customer
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order']> 10), 'loyalty_flag'] = 'Regular Customer'

# locate max order in ords_prods_merge if its < 10 in column loyalty_flag return new customer
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New Customer'
```

```python
# check output
ords_prods_merge.head()
```

```python
# checking frequency
ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

# Question 6

    The team now wants to target different types of spenders in their marketing campaigns. This can be achieved by looking at the prices of the items people are buying. Create a spending flag for each user based on the average price across all their orders using the following criteria:
    If the mean of the prices of products purchased by a user is lower than 10, then flag them as a “Low spender.” If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a “High spender.”

```python
#1.Split the data into groups based on the “user_id” column.

#2.Apply the transform() function on the “prices” column to generate the average price per item for each user.

#3. Create a new column, “user_mean_price,” into which it will place the results of the aggregation.
ords_prods_merge['user_mean_price'] = ords_prods_merge.groupby('user_id')['prices'].transform('mean')
```

```python
#check output
ords_prods_merge['user_mean_price']
```

```python
# locate user mean price in ords_prods_merge if its < 10 in column Spender habit return Low Spender
ords_prods_merge.loc[ords_prods_merge['user_mean_price'] < 10, 'Spender_habit'] = 'Low Spender'

# locate user mean price in ords_prods_merge if its >= 10 in column Sender habit return High Spender
ords_prods_merge.loc[ords_prods_merge['user_mean_price'] >= 10, 'Spender_habit'] = 'High Spender'
```

```python
#output check
ords_prods_merge['Spender_habit'].value_counts(dropna = False)
```

# Question 7

    In order to send relevant notifications to users within the app (for instance, asking users if they want to buy the same item again), the Instacart team wants you to determine frequent versus non-frequent customers. Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column. The criteria for the flag should be as follows:
    
    If the median of “days_since_prior_order” is higher than 20, then the customer should be labeled a “Non-frequent customer.”
    
    If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a “Regular customer.”
    
    If the median is lower than or equal to 10, then the customer should be labeled a “Frequent customer.”


```python
#1.Split the data into groups based on the “user_id” column.

#2.Apply the transform() function on the “days_without_ordering” column to generate the median days without ordering for each user.

#3. Create a new column, “median_days_no_order,” into which it will place the results of the aggregation.
ords_prods_merge['median_days_no_order'] = ords_prods_merge.groupby('user_id')['days_without_ordering'].transform('median')
```

```python
#checking output
ords_prods_merge['median_days_no_order']
```

```python
# locate median days no order in ords_prods_merge if its > 20 in column Customer Type return Non-Frequent Customer
ords_prods_merge.loc[ords_prods_merge['median_days_no_order'] > 20, 'Customer Type'] = 'Non-Frequent Customer'
# locate median days no order in ords_prods_merge if its <= 20 & >10 in column Customer Type return Regular Customer
ords_prods_merge.loc[(ords_prods_merge['median_days_no_order'] <= 20) & (ords_prods_merge['median_days_no_order'] > 10), 'Customer Type'] = 'Regular Customer'
# locate median days no order in ords_prods_merge if its <= 10 in column Customer Type return Frequent Customer
ords_prods_merge.loc[ords_prods_merge['median_days_no_order'] <= 10, 'Customer Type'] = 'Frequent Customer'
```
